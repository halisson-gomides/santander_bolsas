# Modelagem

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(columns='left')
y = df['left']
```

```python
x_train, x_test, y_train, y_test =  train_test_split(x, y, test_size=0.3)
```

```python
x_train.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(x_train, y_train)
```

```python
y_pred = dt.predict(x_test)
```

```python
y_test.value_counts()
```

```python
from sklearn.metrics import plot_confusion_matrix
```

```python
plot_confusion_matrix(dt, x_test, y_test, cmap='Blues')
```

**Explicao da matriz de confuso:**
- 3361 pessoas no vo sair da empresa e de fato o modelo previu que no iro sair
- 65 pessoas no vo sair da empresa, mas o modelo previu que iro sair
- 32 pessoas vo sair da empresa, mas o modelo previu que no iro sair
- 1042 pessoas vo sair da empresa e o modelo previu que iro sair

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
print(classification_report(y_test, y_pred))
```

**Explicao do classification report:**
- _precision_: taxa de acerto sobre o falso-positivo. Nesse contexto: todos funcionrios que o modelo previu que sairiam(1042), mas que no saram da empresa (65) **[olhar para as colunas]**
- _recall_: taxa de previso certa, ou seja, tudo que o modelo previu que realmente aconteceu. **[olhar para as linhas]**
-_f1-score_: media harmnica entre precision e recall

```python
accuracy_score(y_test, y_pred)
```

```python
pd.Series(y_pred).value_counts()
```
